## 6. INTRODUCTION

India, the world's most populous nation since 2023, is undergoing one of the most significant demographic transitions in human history. With a population exceeding 1.4 billion, the country's demographic trajectory has profound implications for economic development, national security, healthcare infrastructure, and strategic resource allocation.

**Demographic transition** refers to the shift from high birth and death rates to low birth and death rates as a society progresses through stages of economic and social development. India is currently in an intermediate-to-advanced stage of this transition, characterised by:

- **Declining fertility rates** — from a Total Fertility Rate (TFR) exceeding 6.0 in the 1950s to approximately 2.0 in recent years.
- **Increasing life expectancy** — from approximately 37 years in 1950 to over 70 years at present.
- **Rapid population aging** — the proportion of elderly (65+) is projected to increase from ~7% to over 20% by 2100.
- **Workforce transformation** — the working-age population (15–64) is approaching its peak, presenting both a demographic dividend opportunity and future workforce sustainability challenges.
- **Changing dependency burden** — the ratio of dependents to working-age population is shifting from youth-heavy to elderly-heavy.

These changes create **long-term demographic risks** related to workforce sustainability, aging population support, healthcare demand, economic productivity, and strategic planning. Understanding and quantifying these risks is essential for evidence-based policy-making and defence preparedness.

This project employs **machine learning techniques** combined with **United Nations WPP 2024 data** to assess demographic risks and forecast population dynamics for India from 1950 to 2100.

## 9. LITERATURE REVIEW

Demographic forecasting and risk assessment have been studied extensively:

- **United Nations (2024):** The World Population Prospects 2024 provides cohort-component projections for 237 countries using assumptions about fertility, mortality, and migration. India is projected to peak around 1.7 billion in the early 2060s.
- **Lee & Carter (1992):** Introduced stochastic mortality forecasting methods that remain influential in actuarial and demographic science.
- **Bongaarts (2009):** Analysed the role of population momentum in sustaining growth after fertility decline.
- **Raftery et al. (2012):** Applied Bayesian probabilistic methods to UN population projections.
- **Prophet (Taylor & Letham, 2018):** Facebook's time-series forecasting tool, widely used for trend-based predictions but limited for complex demographic dynamics.
- **Random Forest (Breiman, 2001):** Ensemble learning method effective for multivariate regression with non-linear relationships.
- **National Commission on Population, India (2020):** Provided domestic population projections aligning with UN estimates.

**Research gap:** Limited application of machine learning with age structure features for demographic risk quantification in the Indian context.

## 12. DATA PREPROCESSING

````python
# ============================================================
# CELL 1: Import Libraries & Load Data
# ============================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import warnings
warnings.filterwarnings('ignore')

# Set global plot style
plt.rcParams.update({
    'figure.figsize': (14, 6),
    'axes.titlesize': 16,
    'axes.labelsize': 13,
    'xtick.labelsize': 11,
    'ytick.labelsize': 11,
    'legend.fontsize': 11,
    'figure.dpi': 120,
    'axes.grid': True,
    'grid.alpha': 0.3,
})
sns.set_style("whitegrid")

print("=" * 60)
print("DRDO-SSPL INTERNSHIP REPORT")
print("Demographic Risk Assessment & Population Forecasting")
print("=" * 60)

# Load Dataset 1: Demographic Indicators
df_demo = pd.read_csv('WPP2024_Demographic_Indicators_Medium.csv.gz', compression='gzip')
india_demo = df_demo[df_demo['Location'] == 'India'].copy()
india_demo = india_demo.sort_values('Time').reset_index(drop=True)

# Load Dataset 2: Population by Age Group
df_age = pd.read_csv('WPP2024_PopulationByAge5GroupSex_Medium.csv.gz', compression='gzip')
india_age = df_age[df_age['Location'] == 'India'].copy()
india_age = india_age.sort_values(['Time', 'AgeGrpStart']).reset_index(drop=True)

print(f"\nDemographic Indicators: {india_demo.shape[0]} rows x {india_demo.shape[1]} cols")
print(f"Time Range: {india_demo['Time'].min()} - {india_demo['Time'].max()}")
print(f"\nAge Structure Data: {india_age.shape[0]} rows x {india_age.shape[1]} cols")
print(f"Age Groups: {india_age['AgeGrp'].nunique()}")

# Select key columns
key_cols = ['Time', 'TPopulation1July', 'PopGrowthRate', 'TFR', 'CBR', 'CDR',
            'LEx', 'MedianAgePop', 'NetMigrations', 'NatChangeRT']
india = india_demo[key_cols].copy()
india.columns = ['Year', 'Population', 'GrowthRate', 'TFR', 'CBR', 'CDR',
                 'LifeExp', 'MedianAge', 'NetMigration', 'NatChangeRate']

# Convert population from thousands to millions
india['Population_M'] = india['Population'] / 1000.0

# ---- Handle Missing Values ----
numeric_cols = india.select_dtypes(include=[np.number]).columns
india[numeric_cols] = india[numeric_cols].interpolate(method='linear', limit_direction='both')
india = india.ffill().bfill()

print("\n--- Key Variables Summary ---")
print(india.describe().round(2).to_string())
print(f"\nMissing values remaining: {india.isnull().sum().sum()}")
print("Data preprocessing complete.")
````

**Interpretation:** The dataset contains 152 annual observations (1950-2101) for India. All key demographic indicators are available. Population values are converted from thousands to millions for interpretability. Any missing values are handled via linear interpolation.

## 14. AGE STRUCTURE ANALYSIS

This section decomposes India's population into three broad age categories:
- **Young Population (0-14):** Pre-working age, dependent on education and child services.
- **Working Population (15-64):** Economically active, productive labour force.
- **Elderly Population (65+):** Post-working age, dependent on pensions and healthcare.

The **Dependency Ratio** is calculated as: *(Young + Elderly) / Working Population*

````python
# ============================================================
# Age Structure Computation
# ============================================================
young_groups = ['0-4', '5-9', '10-14']
working_groups = ['15-19', '20-24', '25-29', '30-34', '35-39', '40-44',
                  '45-49', '50-54', '55-59', '60-64']
elderly_groups = ['65-69', '70-74', '75-79', '80-84', '85-89', '90-94', '95-99', '100+']

age_struct = []
for year in sorted(india_age['Time'].unique()):
    yr_data = india_age[india_age['Time'] == year]
    young = yr_data[yr_data['AgeGrp'].isin(young_groups)]['PopTotal'].sum()
    working = yr_data[yr_data['AgeGrp'].isin(working_groups)]['PopTotal'].sum()
    elderly = yr_data[yr_data['AgeGrp'].isin(elderly_groups)]['PopTotal'].sum()
    total = young + working + elderly
    dep_ratio = (young + elderly) / working if working > 0 else 0
    age_struct.append({
        'Year': year, 'Young': young, 'Working': working, 'Elderly': elderly,
        'Total': total, 'DependencyRatio': dep_ratio,
        'Young_Pct': young/total*100 if total > 0 else 0,
        'Working_Pct': working/total*100 if total > 0 else 0,
        'Elderly_Pct': elderly/total*100 if total > 0 else 0,
    })

age_df = pd.DataFrame(age_struct).sort_values('Year').reset_index(drop=True)

print("Age Structure Summary (Selected Years):")
display_years = [1950, 1975, 2000, 2024, 2050, 2075, 2100]
print(age_df[age_df['Year'].isin(display_years)][
    ['Year', 'Young_Pct', 'Working_Pct', 'Elderly_Pct', 'DependencyRatio']
].to_string(index=False, float_format='%.2f'))
````

````python
# ============================================================
# FIGURE 9: Age Structure Over Time (Stacked Area)
# ============================================================
fig, ax = plt.subplots(figsize=(14, 7))
ax.stackplot(age_df['Year'],
             age_df['Young_Pct'], age_df['Working_Pct'], age_df['Elderly_Pct'],
             labels=['Young (0-14)', 'Working (15-64)', 'Elderly (65+)'],
             colors=['#3498db', '#2ecc71', '#e74c3c'], alpha=0.8)
ax.axvline(x=2024, color='white', linestyle='--', linewidth=2, label='Present')
ax.set_title("Figure 9: Age Structure Composition of India (1950-2100)", fontweight='bold')
ax.set_xlabel("Year")
ax.set_ylabel("Percentage of Total Population (%)")
ax.legend(loc='center right', fontsize=12)
ax.set_xlim(1950, 2100)
ax.set_ylim(0, 100)
plt.tight_layout()
plt.show()

print("\n* Young population share is declining continuously.")
print("* Working-age share peaked and will contract post-2050.")
print("* Elderly share is increasing rapidly.")
````

````python
# ============================================================
# FIGURE 10: Dependency Ratio Over Time
# ============================================================
fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(age_df['Year'], age_df['DependencyRatio'], color='#8e44ad', linewidth=2.5)
ax.fill_between(age_df['Year'], age_df['DependencyRatio'], alpha=0.15, color='#8e44ad')
ax.axvline(x=2024, color='red', linestyle='--', alpha=0.7, label='Present')
ax.set_title("Figure 10: Total Dependency Ratio - India (1950-2100)", fontweight='bold')
ax.set_xlabel("Year")
ax.set_ylabel("Dependency Ratio")
ax.legend()
plt.tight_layout()
plt.show()

print("\n* The dependency ratio fell as the young population shrank (demographic dividend).")
print("* It will rise again as the elderly population expands (aging burden).")
print("* India's 'demographic window' of low dependency is now at or near its optimum.")
````

### Population Momentum and Demographic Dividend

**Demographic Dividend:** India currently benefits from a large working-age population relative to dependents. This 'dividend' fuels economic growth if accompanied by adequate employment, education, and healthcare. The dividend window is closing.

**Population Momentum:** Even after fertility falls below replacement, population continues to grow for decades because large cohorts born during high-fertility periods are still in reproductive years. This explains why India's population peaks well after TFR dropped below 2.1.

## 15. DEMOGRAPHIC RISK ASSESSMENT

Four composite risk indices are developed:

| Index | Formula | Interpretation |
|-------|---------|----------------|
| **Aging Risk Index (ARI)** | `norm(Elderly%) x 0.4 + norm(MedianAge) x 0.3 + norm(LifeExp) x 0.3` | Higher = greater aging burden |
| **Workforce Risk Index (WRI)** | `(1-norm(Working%)) x 0.5 + norm(DepRatio) x 0.3 + (1-norm(GrowthRate)) x 0.2` | Higher = weaker workforce |
| **Population Pressure Index (PPI)** | `norm(Pop) x 0.4 + norm(CBR) x 0.3 + norm(GrowthRate) x 0.3` | Higher = greater pressure |
| **Composite Risk Score (CRS)** | `ARI x 0.35 + WRI x 0.35 + PPI x 0.30` | Overall risk |

**Normalisation:** Min-Max scaling to [0, 1] across 1950-2100.

````python
# ============================================================
# Compute Demographic Risk Indices
# ============================================================
risk_df = india.merge(age_df[['Year', 'Young_Pct', 'Working_Pct', 'Elderly_Pct',
                               'DependencyRatio', 'Young', 'Working', 'Elderly']],
                      on='Year', how='inner')

# Drop any rows with NaN before scaling
risk_df = risk_df.dropna()

scaler = MinMaxScaler()
norm_cols = ['Elderly_Pct', 'MedianAge', 'LifeExp', 'Working_Pct',
             'DependencyRatio', 'GrowthRate', 'Population_M', 'CBR']
for col in norm_cols:
    risk_df[f'n_{col}'] = scaler.fit_transform(risk_df[[col]])

# Aging Risk Index
risk_df['ARI'] = (risk_df['n_Elderly_Pct'] * 0.4 +
                  risk_df['n_MedianAge'] * 0.3 +
                  risk_df['n_LifeExp'] * 0.3)

# Workforce Risk Index
risk_df['WRI'] = ((1 - risk_df['n_Working_Pct']) * 0.5 +
                  risk_df['n_DependencyRatio'] * 0.3 +
                  (1 - risk_df['n_GrowthRate']) * 0.2)

# Population Pressure Index
risk_df['PPI'] = (risk_df['n_Population_M'] * 0.4 +
                  risk_df['n_CBR'] * 0.3 +
                  risk_df['n_GrowthRate'] * 0.3)

# Composite Demographic Risk Score
risk_df['CRS'] = risk_df['ARI'] * 0.35 + risk_df['WRI'] * 0.35 + risk_df['PPI'] * 0.30

print("Risk Indices Summary (Selected Years):")
disp = risk_df[risk_df['Year'].isin([1950, 1975, 2000, 2024, 2050, 2075, 2100])]
print(disp[['Year', 'ARI', 'WRI', 'PPI', 'CRS']].to_string(index=False, float_format='%.3f'))
````

````python
# ============================================================
# FIGURE 11: Demographic Risk Indices Over Time
# ============================================================
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle("Figure 11: Demographic Risk Indices - India (1950-2100)",
             fontsize=16, fontweight='bold')

axes[0,0].plot(risk_df['Year'], risk_df['ARI'], color='#e74c3c', linewidth=2.5)
axes[0,0].fill_between(risk_df['Year'], risk_df['ARI'], alpha=0.15, color='#e74c3c')
axes[0,0].set_title("Aging Risk Index (ARI)", fontweight='bold')
axes[0,0].set_ylabel("Index Value")
axes[0,0].axvline(x=2024, color='gray', linestyle='--', alpha=0.5)

axes[0,1].plot(risk_df['Year'], risk_df['WRI'], color='#2980b9', linewidth=2.5)
axes[0,1].fill_between(risk_df['Year'], risk_df['WRI'], alpha=0.15, color='#2980b9')
axes[0,1].set_title("Workforce Risk Index (WRI)", fontweight='bold')
axes[0,1].set_ylabel("Index Value")
axes[0,1].axvline(x=2024, color='gray', linestyle='--', alpha=0.5)

axes[1,0].plot(risk_df['Year'], risk_df['PPI'], color='#27ae60', linewidth=2.5)
axes[1,0].fill_between(risk_df['Year'], risk_df['PPI'], alpha=0.15, color='#27ae60')
axes[1,0].set_title("Population Pressure Index (PPI)", fontweight='bold')
axes[1,0].set_xlabel("Year"); axes[1,0].set_ylabel("Index Value")
axes[1,0].axvline(x=2024, color='gray', linestyle='--', alpha=0.5)

axes[1,1].plot(risk_df['Year'], risk_df['CRS'], color='#8e44ad', linewidth=2.5)
axes[1,1].fill_between(risk_df['Year'], risk_df['CRS'], alpha=0.15, color='#8e44ad')
axes[1,1].set_title("Composite Demographic Risk Score (CRS)", fontweight='bold')
axes[1,1].set_xlabel("Year"); axes[1,1].set_ylabel("Index Value")
axes[1,1].axvline(x=2024, color='gray', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()

print("\nKey Observations:")
print("* ARI increases monotonically - aging risk is ever-growing.")
print("* WRI rises as workforce share declines and dependency increases.")
print("* PPI peaks around population maximum, then declines.")
print("* CRS captures overall risk trajectory: moderate now, high by 2100.")
````

### 17.1 Model 1: Facebook Prophet (Univariate)

Prophet decomposes a time series into trend, seasonality, and holiday components. It receives Year as `ds` and Population as `y`.

**Strengths:** Simple, automatic changepoint detection.
**Limitations:** Purely univariate, ignores fertility/mortality/age structure. Tends to overestimate by extrapolating recent trends.

````python
# ============================================================
# MODEL 1: Facebook Prophet
# ============================================================
from prophet import Prophet
import logging
logging.getLogger('prophet').setLevel(logging.WARNING)
logging.getLogger('cmdstanpy').setLevel(logging.WARNING)

prophet_df = india[['Year', 'Population_M']].copy()
prophet_df.columns = ['ds', 'y']
prophet_df['ds'] = pd.to_datetime(prophet_df['ds'], format='%Y')

train_prophet = prophet_df[prophet_df['ds'].dt.year <= 2024]

model_prophet = Prophet(yearly_seasonality=False, weekly_seasonality=False,
                        daily_seasonality=False, changepoint_prior_scale=0.05)
model_prophet.fit(train_prophet)

future = model_prophet.make_future_dataframe(periods=76, freq='YS')
forecast_prophet = model_prophet.predict(future)

# FIGURE 12: Prophet Forecast
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(india['Year'], india['Population_M'], 'b-', linewidth=2.5, label='UN Actual/Projected')
ax.plot(forecast_prophet['ds'].dt.year, forecast_prophet['yhat'],
        'r--', linewidth=2, label='Prophet Forecast')
ax.fill_between(forecast_prophet['ds'].dt.year,
                forecast_prophet['yhat_lower'], forecast_prophet['yhat_upper'],
                alpha=0.15, color='red', label='Confidence Interval')
ax.axvline(x=2024, color='green', linestyle='--', alpha=0.7, label='Train/Test Split')
ax.set_title("Figure 12: Prophet Population Forecast vs UN Projection", fontweight='bold')
ax.set_xlabel("Year")
ax.set_ylabel("Population (Millions)")
ax.legend(fontsize=11)
ax.set_xlim(1950, 2100)
plt.tight_layout()
plt.show()

test_years = india[india['Year'] > 2024]['Year'].values
actual_test = india[india['Year'] > 2024]['Population_M'].values
pred_test_p = forecast_prophet[forecast_prophet['ds'].dt.year.isin(test_years)]['yhat'].values[:len(actual_test)]
if len(pred_test_p) == len(actual_test):
    mae_p = mean_absolute_error(actual_test, pred_test_p)
    rmse_p = np.sqrt(mean_squared_error(actual_test, pred_test_p))
    print(f"\nProphet Test MAE: {mae_p:.2f} million")
    print(f"Prophet Test RMSE: {rmse_p:.2f} million")
else:
    mae_p = rmse_p = float('nan')
    print("Prophet test metrics could not be computed (shape mismatch).")
print("\nProphet OVERESTIMATES because it extrapolates the upward trend")
print("without accounting for fertility decline and aging.")
````

**Why Prophet Overestimates:** Prophet extrapolates the historical upward population trend. It has no mechanism to incorporate declining fertility, rising mortality among an aging population, or changing age structure. As a purely trend-based model, it cannot anticipate the demographic transition that causes population to peak and decline.

### 17.2 Model 2: Random Forest (Multivariate)

**Features:** TFR, CBR, CDR, Life Expectancy, Median Age, Net Migration, Growth Rate.

**Strengths:** Captures multivariate non-linear relationships, provides feature importance.
**Limitations:** Cannot extrapolate beyond training data range, causing underestimation.

````python
# ============================================================
# MODEL 2: Random Forest (Demographic Indicators Only)
# ============================================================
features_basic = ['TFR', 'CBR', 'CDR', 'LifeExp', 'MedianAge', 'NetMigration', 'GrowthRate']

# Ensure no NaN in features
india_clean = india.dropna(subset=features_basic + ['Population_M'])

train_rf = india_clean[india_clean['Year'] <= 2024].copy()
test_rf = india_clean[india_clean['Year'] > 2024].copy()

X_train = train_rf[features_basic]
y_train = train_rf['Population_M']
X_test = test_rf[features_basic]
y_test = test_rf['Population_M']

rf_model = RandomForestRegressor(n_estimators=500, max_depth=15, random_state=42, n_jobs=-1)
rf_model.fit(X_train, y_train)

X_all = india_clean[features_basic]
y_pred_all = rf_model.predict(X_all)
y_pred_test = rf_model.predict(X_test)

# FIGURE 13
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(india_clean['Year'], india_clean['Population_M'], 'b-', linewidth=2.5, label='UN Actual/Projected')
ax.plot(india_clean['Year'], y_pred_all, 'g--', linewidth=2, label='RF Prediction')
ax.axvline(x=2024, color='red', linestyle='--', alpha=0.7, label='Train/Test Split')
ax.set_title("Figure 13: Random Forest Population Forecast vs UN Projection", fontweight='bold')
ax.set_xlabel("Year"); ax.set_ylabel("Population (Millions)")
ax.legend(fontsize=11)
ax.set_xlim(1950, 2100)
plt.tight_layout()
plt.show()

mae_rf = mean_absolute_error(y_test, y_pred_test)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2_rf = r2_score(y_test, y_pred_test)
print(f"\nRandom Forest (Basic) Test Metrics:")
print(f"  MAE:  {mae_rf:.2f} million")
print(f"  RMSE: {rmse_rf:.2f} million")
print(f"  R2:   {r2_rf:.4f}")
print("\nRandom Forest UNDERESTIMATES because it cannot extrapolate")
print("beyond the maximum population value seen during training.")
````

**Why Random Forest Underestimates:** RF predictions are bounded by the range of training labels. Since the model was trained on populations up to ~1,440M (2024), it cannot predict values above this. For the period when population exceeds the training maximum, RF systematically underestimates.

### 17.3 Model 3: Random Forest with Age Structure Features

**Additional Features:** Young Population, Working Population, Elderly Population, Dependency Ratio.

**Rationale:** Age structure captures population momentum. A large cohort of young adults guarantees continued births even at low fertility.

````python
# ============================================================
# MODEL 3: Random Forest with Age Structure
# ============================================================
rf_age_df = india.merge(age_df[['Year', 'Young', 'Working', 'Elderly', 'DependencyRatio']],
                        on='Year', how='inner')
rf_age_df = rf_age_df.dropna()

features_age = features_basic + ['Young', 'Working', 'Elderly', 'DependencyRatio']

train_rfa = rf_age_df[rf_age_df['Year'] <= 2024].copy()
test_rfa = rf_age_df[rf_age_df['Year'] > 2024].copy()

X_train_a = train_rfa[features_age]
y_train_a = train_rfa['Population_M']
X_test_a = test_rfa[features_age]
y_test_a = test_rfa['Population_M']

rf_age_model = RandomForestRegressor(n_estimators=500, max_depth=15, random_state=42, n_jobs=-1)
rf_age_model.fit(X_train_a, y_train_a)

y_pred_all_a = rf_age_model.predict(rf_age_df[features_age])

# FIGURE 14
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(rf_age_df['Year'], rf_age_df['Population_M'], 'b-', linewidth=2.5, label='UN Actual/Projected')
ax.plot(rf_age_df['Year'], y_pred_all_a, 'm--', linewidth=2, label='RF + Age Structure')
ax.axvline(x=2024, color='red', linestyle='--', alpha=0.7, label='Train/Test Split')
ax.set_title("Figure 14: RF + Age Structure Forecast vs UN Projection", fontweight='bold')
ax.set_xlabel("Year"); ax.set_ylabel("Population (Millions)")
ax.legend(fontsize=11)
ax.set_xlim(1950, 2100)
plt.tight_layout()
plt.show()

y_pred_test_a = rf_age_model.predict(X_test_a)
mae_rfa = mean_absolute_error(y_test_a, y_pred_test_a)
rmse_rfa = np.sqrt(mean_squared_error(y_test_a, y_pred_test_a))
r2_rfa = r2_score(y_test_a, y_pred_test_a)
print(f"\nRF + Age Structure Test Metrics:")
print(f"  MAE:  {mae_rfa:.2f} million")
print(f"  RMSE: {rmse_rfa:.2f} million")
print(f"  R2:   {r2_rfa:.4f}")
print("\nAge structure features provide information about population momentum.")
````

## 21. DISCUSSION

### Why Demographic Theory Matters
Population dynamics are governed by well-established demographic theories — the Demographic Transition Model, population momentum theory, and cohort-component methodology. Machine learning models that ignore these theoretical foundations produce unreliable long-term forecasts. **Domain knowledge must inform model design.**

### Why Population Forecasting Differs from Ordinary Time-Series Forecasting
Population change is driven by biological (fertility, mortality) and behavioural (migration, marriage patterns) processes that unfold across generations. Unlike financial or weather time-series, population dynamics exhibit:
- **Generational lag:** Changes in fertility take 20–30 years to fully manifest.
- **Structural inertia:** Age structure creates momentum that persists for decades.
- **Non-stationarity:** The underlying demographic regime changes fundamentally over the projection horizon.

Standard time-series methods (ARIMA, Prophet) assume that historical patterns continue. Demographic transitions violate this assumption.

### Why Age Structure is Essential
Age structure determines the number of potential mothers (fertility), the number of elderly (mortality burden), and the size of the workforce (economic capacity). Two countries with identical TFR but different age structures will have vastly different population trajectories. The RF + Age Structure model's improved performance confirms this.

### Why Population Momentum is Important
India's population will continue growing for 30–40 years after TFR dropped below replacement because of the large cohort of young adults currently in reproductive ages. This momentum is invisible to models that consider only aggregate rates. Only models with age-structure awareness can capture it.

### Why Demographic Risk Assessment is Relevant for National Planning
Defence, healthcare, education, pension, and economic policies all depend on population size and structure. The risk indices developed in this study provide a quantitative framework for:
- **Defence planning:** Workforce availability for armed forces recruitment.
- **Healthcare:** Anticipated demand for geriatric care.
- **Economic policy:** Labour supply projections and dependency burden forecasting.
- **Strategic resource allocation:** Infrastructure and service planning for a changing population.

## 22. LIMITATIONS

1. **Data dependency:** All analysis relies on UN medium-variant projections, which are themselves based on assumptions about future fertility, mortality, and migration.
2. **Model extrapolation:** Random Forest cannot extrapolate beyond the training data range, limiting forecast accuracy for unprecedented population levels.
3. **Prophet simplicity:** Prophet treats population as a univariate time-series, ignoring multivariate demographic drivers.
4. **No sub-national analysis:** India has significant state-level demographic variation (e.g., Kerala vs. Bihar) that is not captured in national-level analysis.
5. **Static risk weights:** The risk index weights (0.3, 0.4, etc.) are assigned based on domain knowledge rather than empirically optimised.
6. **No uncertainty quantification:** Except for Prophet's confidence intervals, the study does not provide probabilistic uncertainty bounds on forecasts.
7. **No socioeconomic variables:** Economic growth, urbanisation, education levels, and policy interventions are not explicitly modelled.

## 24. CONCLUSION

This project has conducted a comprehensive demographic risk assessment and population forecasting analysis for India using United Nations World Population Prospects 2024 data and machine learning techniques.

### Executive Summary of Findings

India is transitioning from a phase of **population expansion** toward a phase characterised by:

- **Population aging** — the elderly share (65+) will rise from ~7% to over 20% by 2100.
- **Increasing dependency burden** — as the working-age share contracts and the elderly share expands, the dependency ratio will rise significantly.
- **Workforce sustainability challenges** — the peak of the demographic dividend is imminent; post-peak, the workforce will contract.

**Fertility, mortality, and age structure are the primary drivers** of future demographic change. The Total Fertility Rate has declined to near-replacement level, life expectancy continues to rise, and the median age is increasing rapidly.

### Model Performance

- **Prophet** overestimates future population by extrapolating the historical upward trend without incorporating demographic fundamentals.
- **Random Forest** underestimates future population because tree-based models cannot extrapolate beyond training data bounds.
- **RF + Age Structure** provides improved realism by capturing population momentum through age-structure features.
- **UN cohort-component projections** remain the most reliable benchmark for long-term demographic forecasting.

### Risk Assessment

The **Composite Demographic Risk Score** increases throughout the 21st century, driven primarily by rising aging risk and workforce risk. Population pressure peaks around mid-century before declining as growth turns negative.

### Policy Implications

India must:
1. **Maximise the remaining demographic dividend** through employment generation and skill development.
2. **Prepare for population aging** by strengthening healthcare, pension, and social security systems.
3. **Invest in human capital** to offset declining workforce numbers with increased productivity.
4. **Integrate demographic intelligence** into defence and strategic planning processes.

### Final Statement

*Demographic change is slow but inexorable. The decisions made today will determine whether India successfully navigates the transition from a young, growing population to a mature, potentially declining one. Evidence-based demographic risk assessment, as demonstrated in this study, is essential for informed national planning.*
